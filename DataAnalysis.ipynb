```python
df["time"] = pd.to_numeric(df["time"], errors="coerce")
```

```python
df = df.dropna(subset=["time"])
df["timestamp"] = pd.to_datetime(df["time"], unit="s")
```

```python
df.set_index("timestamp", inplace=True)
```

```python
df.index = pd.to_datetime(df.index)
df = df.sort_index()
```

```python
kitchen_cols = ['Dishwasher [kW]','Microwave [kW]','Fridge [kW]','Kitchen 12 [kW]','Kitchen 14 [kW]','Kitchen 38 [kW]']
available_cols = [c for c in kitchen_cols if c in df.columns]
df_kitchen = df[available_cols]
```

```python
df_kitchen.sum()
```

```python
living_cols = ['Living room [kW]','Home office [kW]']
df_living = df[living_cols]
df_living.sum()
```

```python
heating_cols = ['Furnace 1 [kW]','Furnace 2 [kW]']
df_heating = df[heating_cols]
df_heating.sum()
```

```python
utility_cols = ['Garage door [kW]','Barn [kW]','Well [kW]']
df_utility = df[utility_cols]
df_utility.sum()
```

```python
df["Kitchen Total [kW]"] = df_kitchen.sum(axis=1)
df["Kitchen Total [kW]"].describe()
```

```python
df["Living Total [kW]"] = df_living.sum(axis=1)
df["Living Total [kW]"].describe()
```

```python
df["Heating Total [kW]"] = df_heating.sum(axis=1)
df["Heating Total [kW]"].describe()
```

```python
df["Utility Total [kW]"] = df_utility.sum(axis=1)
df["Utility Total [kW]"].describe()
```

```python
zone_totals = {"Kitchen": df_kitchen.sum().sum(),"Living": df_living.sum().sum(),"Heating": df_heating.sum().sum(),"Utility": df_utility.sum().sum()}
zone_df = pd.DataFrame.from_dict(zone_totals, orient='index', columns=['Total Consumption'])
zone_df.sort_values(by='Total Consumption', ascending=False)
```

```python
df['hour'] = df.index.hour
df['hour'].head()
```

```python
hourly_stats = df.groupby('hour')['use [kW]'].mean()
hourly_stats
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
hourly_stats.plot(kind='bar')
plt.title("Average Energy Consumption by Hour")
plt.xlabel("Hour of the Day")
plt.ylabel("Average kW")
plt.show()
```

```python
kitchen_hourly = df.groupby('hour')['Kitchen Total [kW]'].mean()
kitchen_hourly
```

```python
plt.figure(figsize=(10,5))
kitchen_hourly.plot(kind='bar')
plt.title("Average Kitchen Energy by Hour")
plt.xlabel("Hour")
plt.ylabel("Average kW")
plt.show()
```

```python
mean_power = df['use [kW]'].mean()
mean_power
```

```python
std_power = df['use [kW]'].std()
std_power
```

```python
threshold = mean_power + (2 * std_power)
threshold
```

```python
df['High_Usage'] = df['use [kW]'] > threshold
df['High_Usage'].sum()
```

```python
zones = ["Kitchen Total [kW]", 
         "Living Total [kW]", 
         "Heating Total [kW]", 
         "Utility Total [kW]"]

for zone in zones:
    mean_val = df[zone].mean()
    std_val = df[zone].std()
    threshold = mean_val + (2 * std_val)
    
    df[zone + " High Usage"] = df[zone] > threshold
    
    print(f"\n{zone}")
    print("Mean:", mean_val)
    print("Std Dev:", std_val)
    print("Threshold:", threshold)
    print("High Usage Count:", df[zone + " High Usage"].sum())
```

```python
peak_hour = hourly_stats.idxmax()
peak_value = hourly_stats.max()

print("Peak Hour:", peak_hour)
print("Peak Average Consumption:", peak_value)
```

```python
zone_hourly = df.groupby('hour')[zones].mean()

plt.figure(figsize=(12,6))
zone_hourly.plot()
plt.title("Zone-wise Hourly Consumption Pattern")
plt.xlabel("Hour")
plt.ylabel("Average kW")
plt.show()
```

```python
zone_df['Percentage'] = (
    zone_df['Total Consumption'] / zone_df['Total Consumption'].sum()
) * 100

zone_df.sort_values(by='Percentage', ascending=False)
```
