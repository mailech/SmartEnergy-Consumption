```python
df["time"] = pd.to_numeric(df["time"], errors="coerce")
```

```python
df["time"].isnull().sum()
```

```python
df = df.dropna(subset=["time"])
df["timestamp"] = pd.to_datetime(df["time"], unit="s")
```

```python
df.set_index("timestamp", inplace=True)
```

```python
df.index = pd.to_datetime(df.index)
df = df.sort_index()
```

```python
kitchen_cols = ['Dishwasher [kW]','Microwave [kW]','Fridge [kW]','Kitchen 12 [kW]','Kitchen 14 [kW]','Kitchen 38 [kW]']
available_cols = [c for c in kitchen_cols if c in df.columns]
df_kitchen = df[available_cols]
```

```python
df_kitchen.sum()
```

```python
living_cols = ['Living room [kW]','Home office [kW]']
df_living = df[living_cols]
df_living.sum()
```

```python
heating_cols = ['Furnace 1 [kW]','Furnace 2 [kW]']
df_heating = df[heating_cols]
df_heating.sum()
```

```python
utility_cols = ['Garage door [kW]','Barn [kW]','Well [kW]']
df_utility = df[utility_cols]
df_utility.sum()
```

```python
df["Kitchen Total [kW]"] = df_kitchen.sum(axis=1)
df["Kitchen Total [kW]"].describe()
```

```python
zone_totals = {"Kitchen": df_kitchen.sum().sum(),"Living": df_living.sum().sum(),"Heating": df_heating.sum().sum(),"Utility": df_utility.sum().sum()}
zone_df = pd.DataFrame.from_dict(zone_totals, orient='index', columns=['Total Consumption'])
zone_df.sort_values(by='Total Consumption', ascending=False)
```
