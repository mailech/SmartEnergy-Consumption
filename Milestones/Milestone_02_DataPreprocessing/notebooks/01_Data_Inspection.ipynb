```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))

sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap="viridis")

plt.title("Missing Values Heatmap")

plt.show()
```

```python
df = df.drop_duplicates()
```

```python
df.columns = df.columns.str.strip()
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
small_df = numeric_df.iloc[:, :6]
```

```python
corr = small_df.corr()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(corr, cmap="coolwarm", annot=True, square=True)

plt.title("Correlation Heatmap")

plt.show()
```

```python
energy_col = df.columns[1]

sns.histplot(df[energy_col], bins=40, kde=True)

plt.title("Energy Consumption Distribution")

plt.show()
```

```python
energy_col = df.columns[1]

plt.figure(figsize=(10,5))
plt.plot(df[energy_col])

plt.title("Energy Consumption Trend")

plt.show()
```

```python
# selecting numeric columns
numeric_df = df.select_dtypes(include=['number'])

# correlation matrix
corr_matrix = numeric_df.corr()

# finding correlation with energy column
energy_col = numeric_df.columns[0]

corr_with_energy = corr_matrix[energy_col].sort_values(ascending=False)

corr_with_energy
```
