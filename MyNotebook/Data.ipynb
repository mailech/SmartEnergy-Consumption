```python
df.columns
```

```python
house_cols = ['Home office [kW]','Garage door [kW]','Well [kW]','Living room [kW]','Wine cellar [kW]']
available_cols = [c for c in house_cols if c in df.columns]
```

```python
df_house = df[available_cols]
df_house.sum()
df_house
df_house.notnull()
```

```python
df_house.count()
```

```python
df_house.mean()
df_house.min()
df_house.max()
```

```python
df["total_house_power"] = df[available_cols].sum(axis=1)
df["total_house_power"]
```

```python

df['High_Useage'] = df['total_house_power']>0.00052
df['High_Useage'].describe()
```

```python
df[['total_house_power', 'High_Useage']].sample(5)
```

```python
df['hour'] = df.index.hour
df['hour']
```

```python
hourly_stats = df.groupby('hour')['Well [kW]'].mean()
hourly_stats
plt.figure(figsize=(10,5))
hourly_stats.plot(kind='bar',color='blue')
plt.title("plotting of the Energy consumptions")
plt.xlabel("Hour")
plt.ylabel("KW")
plt.show()
```

```python
hourly_stats = df.groupby('hour')['Garage door [kW]'].mean()
hourly_stats
plt.figure(figsize=(10,5))
hourly_stats.plot(kind='bar',color='red')
plt.title("plotting of the Energy consumptions")
plt.xlabel("Hour")
plt.ylabel("KW")
plt.show()
```

```python
hourly_stats = df.groupby('hour')['Garage door [kW]'].mean()
hourly_stats
plt.figure(figsize=(10,5))
hourly_stats.plot(kind='hist',color='yellow')
plt.title("plotting of the Energy consumptions")
plt.xlabel("Hour")
plt.ylabel("KW")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(hourly_stats.index, hourly_stats.values,color='red')
plt.title("Average Energy Consumption Per Hour")
plt.xlabel("Hour")
plt.ylabel("kW")
plt.show()
```

```python
device_columns = [
    'Dishwasher [kW]',
    'Furnace 1 [kW]',
    'Furnace 2 [kW]',
    'Home office [kW]',
    'Fridge [kW]',
    'Wine cellar [kW]',
    'Garage door [kW]',
    'Kitchen 12 [kW]',
    'Kitchen 14 [kW]',
    'Kitchen 38 [kW]',
    'Barn [kW]',
    'Well [kW]',
    'Microwave [kW]',
    'Living room [kW]'
]
```

```python
df_hourly = df[device_columns].resample('h').sum()
print(df_hourly.head())
```

```python

df_hourly[['Fridge [kW]', 
           'Microwave [kW]', 
           'Dishwasher [kW]']].plot(figsize=(10,5))

plt.title("Hourly Device wise Energy Consumption")
plt.xlabel("Time")
plt.ylabel("Energy in kW")
plt.show()
```

```python
total_consumption = df[device_columns].sum().sort_values(ascending=False)
print(total_consumption)
```

```python
Top5 = total_consumption.head(5)
print("Top 5 High Energy Consuming Devices are:")
print(Top5)
```

```python
import matplotlib.pyplot as plt
Top5.plot(kind='bar', figsize=(8,5),color='red')
plt.title("Top 5 Highest Energy Consuming Devices")
plt.ylabel("Total Energy Consumption (kW)")
plt.xticks(rotation=45)
plt.show()
```
